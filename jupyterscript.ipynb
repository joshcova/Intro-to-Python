Let us now focus on a subset of countries: We are only interested in Germany, Italy, France and Sweden. 
Let us define a list first containing these countries and then subset the dataframe df by excluding all other countries

```python
# There are a lot of countries in our dataset. How many? 
df["Country Name"].unique() # prints out unique value (i.e. the name of the countries)
df["Country Name"].nunique() # print the number of unique values (i.e. the number of countries)
```

```python
countries = ["Germany", "Italy", "France", "Sweden"] # We select the countries we are interested in
select_df = df.loc[df["Country Name"].isin(countries)]
# We create a new df called select_df
# We locate through the .loc method the rows where the Country name equals the values that we have specified in the countries list
```

```python
select_df["Country Name"].unique() # Let us check the unique countries to see that it worked
```

Let us now create a pretty line plot with the seaborn package (sns) that show us how public sector compensation as a share of GDP 
has changed in time.

```python
sns.lineplot(data= select_df,
             x="Year",
             y="Value", 
             hue = "Country Name")
```
